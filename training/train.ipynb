```python
from ultralytics import YOLO
```

```python
model = YOLO('/kaggle/input/datasets/ayushhalpati/nut-screw-model/final_best.pt')
```

```python
model.train(
    data='/kaggle/input/datasets/ayushhalpati/addbackgroundfinaldatasetforneno/final_v5_balanced_output/data.yaml',
    epochs=40,
    imgsz=640,
    batch=16,
    lr0=0.0003,
    device=0,          # GPU
    workers=2,

    # Augmentations
    mosaic=0.3,
    mixup=0.1,
    copy_paste=0.5,

    # Optimization
    patience=10,
    optimizer='AdamW',

    # Save
    project='yolo_training',
    name='final_run',
    exist_ok=True
)
```

```python
model = YOLO('/kaggle/working/runs/segment/yolo_training/final_run/weights/best.pt')
model.export(format="onnx", imgsz=416, half=True)
```
